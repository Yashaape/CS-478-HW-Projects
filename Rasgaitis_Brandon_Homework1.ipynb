Activation Function: Identity
g(x)=x

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.01)
y = x 
plt.plot(x,y)
plt.title('Identity')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function: Binary Step
g(x) = {0 if x < 0},
       {1 if x >= 0}




```python
x = np.arange(-5, 5, 0.01)
y = np.heaviside(x,1)
plt.plot(x,y)
plt.title('Binary Step')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function: Sigmoid g(x)= 1 / 1 + e^-x

```python
x = np.arange(-5, 5, 0.01)
y = 1 / (1+np.exp(-x))
plt.plot(x,y)
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function: Hyperbolic Tangent(tanh)
g(x) = (1-exp(-2x))/(1+exp(-2x))

```python
x = np.arange(-5, 5, 0.01)
y = np.tanh(x)
plt.plot(x,y)
plt.title('Tanh')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function:Rectified Linear Unit (ReLU)
g(x) = It returns zero if the input is less than zero otherwise it returns the given input

```python
x = np.arange(-5, 5, 0.01)
y = []
for i in x:
  if(i < 0):
    y.append(0)
  else: 
    y.append(i)
plt.plot(x,y)
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function: Gaussian Error Linear Unity (GeLU)
g(x) = (1/2)x(1 + erf(x/sqrt(2))

```python
import tensorflow as tf

x = np.arange(-5, 5, 0.01)
y = 0.5 * (1.0 + tf.math.erf(x / tf.sqrt(2.0)))
y = x * y
plt.plot(x,y)
plt.title('GeLU')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function: Softplus 
g(x) = ln(1+e^x) 

```python
x = np.arange(-5, 5, 0.01)
y = tf.nn.softplus(x)
with tf.Session() as sess:
  plt.plot(x, sess.run(y))
  plt.title('Softplus')
  plt.xlabel('x')
  plt.ylabel('g(x)');
  plt.grid()
  plt.show()
```

Activation Function: Exponential linear unit(ELU)
g(x) = if `x` > 0, output would be `x`
       if `x` < 0, output would be calculated by    function `alpha * (exp(x)-1)`, where `alpha` is a    scalar.


```python
def elu(x):
  y=[]
  for i in x:
      if i<0:
          y.append(0.5*(np.exp(i) -1))
      else:
          y.append(i)
  return y

x = np.arange(-5, 5, 0.01)
y = elu(x)

plt.plot(x,y)
plt.title('ELU')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

Activation Function: Scaled exponential linear unit(SELU)
SELU(x)=λ{x if x>0 }, {αe^x−α if x≤0}

```python
λ = 1.0507
α = 1.6732

def SELU(x):
  if x > 0:
      return λ*x
  return λ*α*(np.exp(x) - 1)

x = np.arange(-5, 5, 0.01)
y = []
for i in x:
  y.append(SELU(i))
  
plt.plot(x,y)
plt.title('SELU')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

Activation Function: Leaky rectified linear unit
(Leaky ReLU)
g(x) = {0.01x if x < 0}, {x if x >= 0}

```python
def leaky_ReLU(x):
  y = [max(0.05*value,value) for value in x]
  return np.array(y, dtype=float)

x = np.arange(-5, 5, 0.01)
y = leaky_ReLU(x)

plt.plot(x,y)
plt.title('Leaky ReLU')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

Activation Function: Parametric rectified lienar unit(PReLU)
g(x) = {αx if x < 0}, {x if x >= 0}

```python
def prelu(x,a=0.1):
   return np.maximum(a*x,x)

x = np.arange(-5, 5, 0.01)
xinput = [x for x in x]
y = [prelu(x,a=0.1) for x in xinput]

plt.plot(x,y)
plt.title('PReLU')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

#Might be har to notice, but the PReLU graph has a lower -g(x) value
```

Activation Function: Sigmoid linear unit(SiLU)
g(x) = x / (1 + e^-x)

```python
def swish(x):
    s= x / (1.0+ np.exp(-x))
    return s

x = np.arange(-5, 5, 0.01)
y = swish(x)
plt.plot(x,y)
plt.title('SiLU')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

Activation Function: Gaussian 
g(x) = e^-x^2

```python
import math 
def gaussian(x):
  return np.exp(-1*x*x)

x = np.arange(-5, 5, 0.01)
y = gaussian(x)
plt.plot(x,y)
plt.title('Gaussian')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```
